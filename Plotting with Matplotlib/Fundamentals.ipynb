```python
data = {'a': np.arange(50),
        'b': np.random.randint(0,50,50),
        'c': np.random.randn(50)}
data['b'] = data['a'] + 10 * data['c']
data['d'] = np.abs(data['c']) * 100
data
```

```python
plt.scatter('a', 'b', c = 'c', s = 'd', data = data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()

```

```python
names = ['group a ', 'group b', 'group c']
values = [1,10,100]
plt.figure(figsize=(9,3))
plt.subplot(131)
plt.bar(names,values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('categorical plot')
plt.show()
```

```python
import ssl
import urllib.request
ssl._create_default_https_context = ssl._create_unverified_context

import seaborn as sns
import pandas as pd
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
type(iris)
iris.describe()
```

```python
plt.scatter(iris.sepal_length, iris.sepal_width)
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
plt.title('Iris Sepal Length vs Width')
plt.show()
```

```python
plt.boxplot(x='petal_length', data=iris)
plt.title('Iris Sepal Length by Species')
plt.show()

```

```python
plt.hist(iris['petal_width'], bins = 20)
plt.title('Iris Sepal Length Distribution')
plt.show()
```

```python
species_counter = iris['species'].value_counts()
plt.pie(species_counter, labels=species_counter.index, autopct='%1.1f%%', startangle=140)
plt.title('Iris Species Distribution')
plt.show()
```

```python
sns.set_theme(style="ticks")
sns.pairplot(iris, hue="species")
plt.show
```

```python
sns.set_palette(sns.color_palette("husl", 3))
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 6))
sns.boxplot(x='species', y='sepal_length', data=iris, ax=ax[0, 0])
sns.boxplot(x='species', y='sepal_width', data=iris, ax=ax[0, 1])
sns.boxplot(x='species', y='petal_length', data=iris, ax=ax[1, 0])
sns.boxplot(x='species', y='petal_width', data=iris, ax=ax[1, 1])
plt.tight_layout()
plt.show() 

```

```python
sns.set_style("whitegrid")
sns.violinplot(x='species', y='petal_length', data=iris)
plt.title('Iris Petal Length by Species')
plt.show()
```

```python
sns.set_style("whitegrid")
sns.lmplot(x='petal_length', y='sepal_length', hue='species', data=iris)
plt.title('Iris Sepal Length vs Petal Length by Species')
plt.show()
```

```python
sns.heatmap(iris.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.show()
```
